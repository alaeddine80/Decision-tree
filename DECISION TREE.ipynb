```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
#importing relevant libraries
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

 # Reading the dataset

```python
A=pd.read_csv('titanic1.csv')
A
```

```python
A1=A.filter(['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','C','Q','S'])
A1
```

# Splitting Data

```python
x=A1[['Pclass','Sex','Age','SibSp','Parch','Fare','C','Q','S']]
y=A1['Survived']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)
x_train, x_test, y_train, y_test 
```

# Applying the decision tree

# Decision Tree

```python
clf= tree.DecisionTreeClassifier()  
clf.fit(x_train, y_train)   #fitting our model
y_pred=clf.predict(x_test)   # evaluating our model
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

# plotting the decision tree

```python
dot_data =tree.export_graphviz(clf,out_file=None)
graph=graphviz.Source(dot_data)
graph.render('A1')
graph

```

# changing the parameters

```python
x1=A1[['Pclass','Sex','Age','SibSp','Parch','Fare','C','Q','S']]
y1=A1['Survived']
x1_train, x1_test, y1_train, y1_test = train_test_split(x1,y1, test_size=0.35,random_state=20)
x1_train, x1_test, y1_train, y1_test 
```

```python
clf= tree.DecisionTreeClassifier()  
clf.fit(x1_train, y1_train)   #fitting our model
y1_pred=clf.predict(x1_test)   # evaluating our model
print("score:{}".format(accuracy_score(y1_test, y1_pred)))
```

```python
#After changing the decision tree parameters the accuracy remain the same
```

# Random forest

```python
clf=RandomForestClassifier(n_estimators=150)  #Creating a random forest with 100 decision trees
clf.fit(x_train, y_train)  #Training our model
y_pred=clf.predict(x_test)  #testing our model
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))  
```

# changing the number of estimator

```python
clf=RandomForestClassifier(n_estimators=200)  #Creating a random forest with 100 decision trees
clf.fit(x_train, y_train)  #Training our model
y_pred=clf.predict(x_test)  #testing our model
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))  
```
